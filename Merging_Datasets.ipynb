```python
import pandas as pd
```

```python
# Load original datasets (unmodified)
df_ChronicRD = pd.read_csv("Chronic respiratory .csv")
df_ChronicRD_hbm = pd.read_csv("Chronic respiratory desease W:HBM.csv")
```

```python
#filter to have only the number of deaths
df_ChronicRD = df_ChronicRD[df_ChronicRD['measure_name'] == 'Deaths']
df_ChronicRD_hbm = df_ChronicRD_hbm[df_ChronicRD_hbm['measure_name'] == 'Deaths']
```

```python
#Standardize the location names
country_name_mapping = {
    "Federal Republic of Nigeria": "Nigeria",
    "Federative Republic of Brazil": "Brazil",
    "Kingdom of the Netherlands": "Netherlands",
    "Republic of India": "India"
}

df_ChronicRD['location_name'] = df_ChronicRD['location_name'].replace(country_name_mapping)
```

```python
# Define 'Under 40' age bands
under_40 = [
    '10-14 years', '15-19 years', '20-24 years',
    '25-29 years', '30-34 years', '35-39 years'
]

# Create 'age_group' column
df_ChronicRD['age_group'] = df_ChronicRD['age_name'].apply(lambda x: 'Under 40' if x in under_40 else '40 and above')
df_ChronicRD_hbm['age_group'] = df_ChronicRD_hbm['age_name'].apply(lambda x: 'Under 40' if x in under_40 else '40 and above')
```

```python
#group by measure name
grouped_ChronicRD = df_ChronicRD.groupby(
    ['location_name', 'year', 'age_group', 'sex_name']
)['val'].sum().reset_index()

grouped_ChronicRD_hbm = df_ChronicRD_hbm.groupby(
    ['location_name', 'year', 'age_group', 'sex_name']
)['val'].sum().reset_index()
```

```python
#merge the two datasets
df_chronic_merged = pd.merge(
    grouped_ChronicRD,
    grouped_ChronicRD_hbm,
    on=['location_name', 'year', 'age_group', 'sex_name'],
    suffixes=('_total', '_hbm')
)
```

```python
#create a column for the percentage of deaths due to HBM
df_chronic_merged['percent_hbm'] = (
    df_chronic_merged['val_hbm'] / df_chronic_merged['val_total']
) * 100

```

```python
# Create formatted versions of total and HBM death columns
df_chronic_merged['val_total_formatted'] = df_chronic_merged['val_total'].apply(lambda x: f"{int(round(x)):,}")
df_chronic_merged['val_hbm_formatted'] = df_chronic_merged['val_hbm'].apply(lambda x: f"{int(round(x)):,}")
```

```python
import pandas as pd

# Load your two CSV files
stroke_df = pd.read_csv("Stroke.csv")
stroke_hbm_df = pd.read_csv("Stroke_hbm.csv")

# Filter to only include rows where measure is 'Deaths'
stroke_df_deaths = stroke_df[stroke_df['measure_name'] == 'Deaths'].copy()
stroke_hbm_df_deaths = stroke_hbm_df[stroke_hbm_df['measure_name'] == 'Deaths'].copy()

# Create a function to classify age into groups
def age_group(age_name):
    under_40_keywords = ['10', '15', '20', '25', '30', '35']
    for keyword in under_40_keywords:
        if keyword in age_name:
            return 'Under 40'
    return '40 and above'

# Apply the age group
stroke_df_deaths['age_group'] = stroke_df_deaths['age_name'].apply(age_group)
stroke_hbm_df_deaths['age_group'] = stroke_hbm_df_deaths['age_name'].apply(age_group)

# Group and aggregate total stroke deaths
stroke_total_grouped = (
    stroke_df_deaths
    .groupby(['location_name', 'year', 'age_group', 'sex_name'])['val']
    .sum()
    .reset_index()
    .rename(columns={'val': 'val_total'})
)

# Group and aggregate high BMI stroke deaths
stroke_hbm_grouped = (
    stroke_hbm_df_deaths
    .groupby(['location_name', 'year', 'age_group', 'sex_name'])['val']
    .sum()
    .reset_index()
    .rename(columns={'val': 'val_hbm'})
)

# Merge the two grouped datasets
merged = pd.merge(
    stroke_total_grouped,
    stroke_hbm_grouped,
    on=['location_name', 'year', 'age_group', 'sex_name'],
    how='left'
)

# Calculate percent_hbm
merged['percent_hbm'] = (merged['val_hbm'] / merged['val_total']) * 100

# Calculate percent_change_from_2015
def calc_percent_change(group):
    base_val = group[group['year'] == 2015]['val_total']
    if not base_val.empty:
        group['percent_change_from_2015'] = ((group['val_total'] - base_val.values[0]) / base_val.values[0]) * 100
    else:
        group['percent_change_from_2015'] = pd.NA
    return group

merged = merged.groupby(['location_name', 'age_group', 'sex_name'], group_keys=False).apply(calc_percent_change)

# Optional: Save to CSV
merged.to_csv("stroke_merged_output.csv", index=False)
```

```python
# Dictionary of name replacements
country_name_map = {
    "Federal Republic of Nigeria": "Nigeria",
    "Federative Republic of Brazil": "Brazil",
    "Kingdom of the Netherlands": "Netherlands",
    "Republic of India": "India"
}

# Replace the names
merged['location_name'] = merged['location_name'].replace(country_name_map)

# Optional: Save the updated version
merged.to_csv("stroke_merged_output.csv", index=False)
```

```python
# Load each disease-specific merged file
alzheimers = pd.read_csv("Alzheimers_merged.csv")
diabetes = pd.read_csv("diabetes_merged.csv")
ischemic = pd.read_csv("Ischemic_merged.csv")
respiratory = pd.read_csv("Respiratory_merged.csv")
stroke = pd.read_csv("stroke_merged_output.csv")

# Add a 'disease' column to each DataFrame
alzheimers['disease'] = 'Alzheimers'
diabetes['disease'] = 'Diabetes'
ischemic['disease'] = 'Ischemic Heart Disease'
respiratory['disease'] = 'Chronic Respiratory Disease'
stroke['disease'] = 'Stroke'

# Combine all the datasets
all_diseases_df = pd.concat([alzheimers, diabetes, ischemic, respiratory, stroke], ignore_index=True)

#save to CSV
all_diseases_df.to_csv("all_diseases_merged.csv", index=False)
```
